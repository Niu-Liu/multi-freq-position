Of course, X does not follow a Rayleigh-shape distribution of unit variance.

```python
lim = 1000

mask0 = r2o["nor_sep_g1"] < lim
r2o0 = r2o[mask0]
loc0, scale0 = rayleigh.fit(r2o0["nor_sep_g1"], floc=0)
print("There are {:d} sources with normalized separation X-G > {:d}".format(len(r2o)-len(r2o0), lim))
print("The standard deviation is {:.2f}".format(scale0))

mask1 = r2o["nor_sep_k_g1"] < lim
r2o1 = r2o[mask1]
loc1, scale1 = rayleigh.fit(r2o0["nor_sep_k_g1"], floc=0)
print("There are {:d} sources with normalized separation K-G > {:d}".format(len(r2o)-len(r2o1), lim))
print("The standard deviation is {:.2f}".format(scale1))

mask2 = r2o["nor_sep_ka_g1"] < lim
r2o2 = r2o[mask2]
loc2, scale2 = rayleigh.fit(r2o0["nor_sep_ka_g1"], floc=0)
print("There are {:d} sources with normalized separation Ka-G > {:d}".format(len(r2o)-len(r2o2), lim))
print("The standard deviation is {:.2f}".format(scale2))
```

Set a limit on the X to see the distribution of X.

```python
xlim = np.arange(1, 10, 0.25)

sounb0 = np.zeros_like(xlim)
sounb1 = np.zeros_like(xlim)
sounb2 = np.zeros_like(xlim)
std0 = np.zeros_like(xlim)
std1 = np.zeros_like(xlim)
std2 = np.zeros_like(xlim)

for i, lim in enumerate(xlim):

    # X - G
    mask0 = r2o["nor_sep_g1"] < lim
    r2o0 = r2o[mask0]
    loc0, scale0 = rayleigh.fit(r2o0["nor_sep_g1"], floc=0)
    sounb0[i] = len(r2o0)
    std0[i] = scale0

    # K - G
    mask1 = r2o["nor_sep_k_g1"] < lim
    r2o1 = r2o[mask1]
    loc1, scale1 = rayleigh.fit(r2o0["nor_sep_k_g1"], floc=0)
    sounb1[i] = len(r2o1)
    std1[i] = scale1

    # Ka - G
    mask2 = r2o["nor_sep_ka_g1"] < lim
    r2o2 = r2o[mask2]
    loc2, scale2 = rayleigh.fit(r2o0["nor_sep_ka_g1"], floc=0)
    sounb2[i] = len(r2o2)
    std2[i] = scale2
```

I started with the limit of $\rho$ to be 20 mas, but it shows that there are few sources with $\rho$> 10 mas.

I want to check how limit on the radio-to-optical offset on one band affects the distribution of other bands.

```python
# X - G

rholim = np.arange(0.1, 10.1, 0.2)
sounb0 = np.zeros_like(rholim)
sounb1 = np.zeros_like(rholim)
sounb2 = np.zeros_like(rholim)
std0 = np.zeros_like(rholim)
std1 = np.zeros_like(rholim)
std2 = np.zeros_like(rholim)

for i, lim in enumerate(rholim):

    mask0 = r2o["ang_sep_g1"] < lim
    r2o0 = r2o[mask0]

    loc0, scale0 = rayleigh.fit(r2o0["nor_sep_g1"], floc=0)
    sounb0[i] = len(r2o0)
    std0[i] = scale0

    loc1, scale1 = rayleigh.fit(r2o0["nor_sep_k_g1"], floc=0)
    sounb1[i] = len(r2o1)
    std1[i] = scale1

    loc2, scale2 = rayleigh.fit(r2o0["nor_sep_ka_g1"], floc=0)
    sounb2[i] = len(r2o2)
    std2[i] = scale2
```
