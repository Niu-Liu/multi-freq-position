```python
from astropy.table import Table, join, Column
from astropy import units as u
import astropy.coordinates as coord
from astropy.coordinates import SkyCoord
import numpy as np

import matplotlib.pyplot as plt
from matplotlib.ticker import MultipleLocator, NullFormatter
%matplotlib inline
# %config InlineBackend.figure_format = "svg"
```

```python
r2o= Table.read("../data/common-source-position-wrt-gaia.fits")
```

```python
# Gaia DR2 photometric information
from my_progs.catalog.read_gaia import read_dr2_iers

gdr2 = read_dr2_iers()

gdr2.keep_columns(["iers_name", "source_id",
                   "ra", "ra_err", "dec", "dec_err", "pos_err",
                   "parallax", "parallax_err",
                   "pmra", "pmra_err", "pmdec", "pmdec_err",
                   "astrometric_n_obs_al", "astrometric_matched_observations",
                   "phot_g_mean_mag", "phot_bp_mean_mag", "phot_rp_mean_mag", "bp_rp"])

gdr2.rename_columns(["astrometric_n_obs_al", "astrometric_matched_observations",
                     "phot_g_mean_mag", "phot_bp_mean_mag", "phot_rp_mean_mag"],
                    ["n_obs_al", "n_matched_obs", "g_mag", "bp_mag", "rp_mag"])

r2o = join(r2o, gdr2, keys="iers_name", table_names=["", "g"])
```

# Angular separation $\rho$

This plot shows that the VLBI position precision has less correlation with the $G$-magnitude.

```python
# Distribution of angular separation
fig, (ax0, ax1, ax2) = plt.subplots(figsize=(6, 8), nrows=3, sharex=True, sharey=True)

ax0.plot(r2o["g_mag"], r2o["ang_sep_x_all2"], ".", ms=3)
ax1.plot(r2o["g_mag"], r2o["ang_sep_k_all2"], ".", ms=3)
ax2.plot(r2o["g_mag"], r2o["ang_sep_ka_all2"], ".", ms=3)

# ax0.plot(dat_binned["g_mag"], dat_binned)


ax0.set_xlim([12.5, 21.5])

ax0.set_ylabel("$\\rho$ (S/X$-$Gaia, mas)")
ax1.set_ylabel("$\\rho$ (K$-$Gaia, mas)")
ax2.set_ylabel("$\\rho$ (X/Ka$-$Gaia, mas)")
ax2.set_xlabel("$G$ (mag)")

ax0.set_yscale("log")
ax1.set_yscale("log")
ax2.set_yscale("log")

ax0.xaxis.set_ticks_position("both")
ax0.yaxis.set_ticks_position("both")
ax1.xaxis.set_ticks_position("both")
ax1.yaxis.set_ticks_position("both")
ax2.xaxis.set_ticks_position("both")
ax2.yaxis.set_ticks_position("both")

plt.subplots_adjust()
plt.tight_layout()
```

```python
# Distribution of angular separation
fig, (ax0, ax1, ax2) = plt.subplots(figsize=(6, 8), nrows=3, sharex=True, sharey=True)

ax0.plot(r2o["g_mag"], r2o["nor_sep_x_all2"], ".", ms=3)
ax1.plot(r2o["g_mag"], r2o["nor_sep_k_all2"], ".", ms=3)
ax2.plot(r2o["g_mag"], r2o["nor_sep_ka_all2"], ".", ms=3)

ax0.set_xlim([12.5, 21.5])

ax0.set_ylabel("$X$ (S/X$-$Gaia)")
ax1.set_ylabel("$X$ (K$-$Gaia)")
ax2.set_ylabel("$X$ (X/Ka$-$Gaia)")
ax2.set_xlabel("$G$ (mag)")

ax0.set_yscale("log")
ax1.set_yscale("log")
ax2.set_yscale("log")

ax0.xaxis.set_ticks_position("both")
ax0.yaxis.set_ticks_position("both")
ax1.xaxis.set_ticks_position("both")
ax1.yaxis.set_ticks_position("both")
ax2.xaxis.set_ticks_position("both")
ax2.yaxis.set_ticks_position("both")

plt.subplots_adjust()
plt.tight_layout()
```

Generally the sample shows that faint sources tends to have a large radio-to-optical distance but smaller $X$.

```python
# Distribution of angular separation
fig, (ax0, ax1) = plt.subplots(figsize=(4, 6), nrows=2, sharex=True, sharey=True)

ax0.plot(r2o["ang_sep_x_all2"], r2o["ang_sep_k_all2"], ".", ms=2)
ax1.plot(r2o["ang_sep_x_all2"], r2o["ang_sep_ka_all2"], ".", ms=2)


ax1.set_xlabel("$\\rho$ (S/X$-$Gaia, mas)")
ax0.set_ylabel("$\\rho$ (K$-$Gaia, mas)")
ax1.set_ylabel("$\\rho$ (X/Ka$-$Gaia, mas)")

ax0.set_xlim([0.01, 20])
ax0.set_ylim([0.01, 20])

ax0.set_xscale("log")
ax0.set_yscale("log")
ax1.set_xscale("log")
ax1.set_yscale("log")

x = np.linspace(0.01, 20, 100)
ax0.plot(x, x, lw=0.5)
ax1.plot(x, x, lw=0.5)

ax0.xaxis.set_ticks_position("both")
ax0.yaxis.set_ticks_position("both")
ax1.xaxis.set_ticks_position("both")
ax1.yaxis.set_ticks_position("both")

plt.subplots_adjust()
plt.tight_layout()
```
