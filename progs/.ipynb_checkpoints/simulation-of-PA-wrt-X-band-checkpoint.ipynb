In this notebook, I compared the "real" (in the catalog) and simulation positions.

The simulation positions were obtained by considering a perturbation following the error ellipse on the positions in the catalog.

Then I calculated the distances and position angles of $K$-, $Ka$-band, and \textit{Gaia} positions with referred to the $X$-band one.

For individual source, we adopted the peak value of the simulated sample (10000 points) as the estimated value for distance and position angle.

```python
from astropy.table import Table, join, Column
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import MultipleLocator, NullFormatter
import time
from scipy import interpolate
%matplotlib inline

# My module
from calc_pa import pa_calc, inclu_angle_calc, wrap_angle
from my_progs.catalog.read_icrf import read_icrf3
from my_progs.catalog.read_gaia import read_dr2_iers
from error_ellipse import search_for_pa_peak, search_for_rho_peak
```

```python
icrf3x = read_icrf3(wv="sx")
icrf3x.keep_columns(["iers_name", "ra_dec_corr"])
icrf3x.rename_column("ra_dec_corr", "ra_dec_corr_x")

icrf3k = read_icrf3(wv="k")
icrf3k.keep_columns(["iers_name", "ra_dec_corr"])
icrf3k.rename_column("ra_dec_corr", "ra_dec_corr_k")

icrf3ka = read_icrf3(wv="xka")
icrf3ka.keep_columns(["iers_name", "ra_dec_corr"])
icrf3ka.rename_column("ra_dec_corr", "ra_dec_corr_ka")

gdr2 = read_dr2_iers()
gdr2.keep_columns(["iers_name", "ra_dec_corr"])
gdr2.rename_column("ra_dec_corr", "ra_dec_corr_g")
```

```python
comsou = Table.read("../data/common-source-position.fits")

comsou = join(comsou, icrf3x, keys="iers_name")
comsou = join(comsou, icrf3k, keys="iers_name")
comsou = join(comsou, icrf3ka, keys="iers_name")
comsou = join(comsou, gdr2, keys="iers_name")
```

```python
data = Table.read("../logs/sim_pa_and_offset.log", format="ascii",
                  names=["iers_name",
                         "ang_sep_k_sim", "ang_sep_k_sim1", "ang_sep_k_sim2",
                         "pa_k_sim", "pa_k_sim1", "pa_k_sim2",
                         "ang_sep_ka_sim", "ang_sep_ka_sim1", "ang_sep_ka_sim2",
                         "pa_ka_sim", "pa_ka_sim1", "pa_ka_sim2",
                         "ang_sep_g_sim", "ang_sep_g_sim1", "ang_sep_g_sim2",
                         "pa_g_sim", "pa_g_sim1", "pa_g_sim2"])

comsou = join(comsou, data, keys="iers_name")
```

Now we see how the error ellipse bias the offset and PA.

This plot compares the simulated position distances with the original ones.

```python
fig, (ax0, ax1, ax2) = plt.subplots(figsize=(4, 9), nrows=3, sharex=True, sharey=True)

x = np.linspace(0.01, 25, 20)

ax0.plot(x, x, "r-", lw=0.5)
ax1.plot(x, x, "r-", lw=0.5)
ax2.plot(x, x, "r-", lw=0.5)

ax0.plot(comsou["ang_sep_k_ccl2"], comsou["ang_sep_k_sim"], ".", ms=1, lw=0.5, markerfacecolor="None")
ax1.plot(comsou["ang_sep_ka_ccl2"], comsou["ang_sep_ka_sim"], ".", ms=1, lw=0.5, markerfacecolor="None")
ax2.plot(comsou["ang_sep_g_ccl2"], comsou["ang_sep_g_sim"], ".", ms=1, lw=0.5, markerfacecolor="None")

ax0.set_xscale("log")
ax0.set_yscale("log")
ax0.axis([0.01, 25, 0.01, 25], "square")
ax2.set_xlabel("Original distance (mas)")
ax0.set_ylabel("Simulated distance (mas)")
ax1.set_ylabel("Simulated distance (mas)")
ax2.set_ylabel("Simulated distance (mas)")


ax0.text(0.5, 10, "$K-X$", fontsize=12)
ax1.text(0.5, 10, "$Ka-X$", fontsize=12)
ax2.text(0.5, 10, "${\it Gaia}-X$", fontsize=12)

plt.tight_layout()
```

We could see a tail when original distance is less than 0.1 mas (~0.2 mas for $X$-to-Gaia), suggesting a budget of 0.1 mas on the position agreement for extra-galactic sources.

To understand this budget, I then looked at the sub-sample of original distance < 0.2 mas.

See from their positional uncertainties (semi-major axis of the error ellipse), these sources usually have a precise position at $X$-band, whose precision is close to the noise floor of ICRF3.

But the formal uncertainties of their positions at other bands is aound 0.1 mas or large (Gaia).

As a result, the simulated position distance is likely to reflect the combined position formal uncertainties between $X$-band and other three bands.

Anyway, I moved forward to compute the mean histogram of PA of X-to-K, X-to-Ka, and X-to-Gaia vector.

The red curve is the mean histogram, with layers in different colors indicating the 1-$\sigma$ to 3-$\sigma$ confidence intervals.

The peaks at 180$^\circ$ for X-to-K and X-to-Ka could be recovered.

So do the asymmetry distribution of X-to-Gaia.

```python
pa_k_sim = Table.read("../data/pa_k_sim_hist.dat", format="ascii")
pa_ka_sim = Table.read("../data/pa_ka_sim_hist.dat", format="ascii")
pa_g_sim = Table.read("../data/pa_gaia_sim_hist.dat", format="ascii")
```

```python
cols = pa_k_sim.colnames

binmean_k = np.zeros(len(cols))
binstd_k = np.zeros(len(cols))
binmean_ka = np.zeros(len(cols))
binstd_ka = np.zeros(len(cols))
binmean_g = np.zeros(len(cols))
binstd_g = np.zeros(len(cols))

for i, col in enumerate(cols):
    binmean_k[i] = np.mean(pa_k_sim[col])
    binstd_k[i] = np.std(pa_k_sim[col])
    binmean_ka[i] = np.mean(pa_ka_sim[col])
    binstd_ka[i] = np.std(pa_ka_sim[col])
    binmean_g[i] = np.mean(pa_g_sim[col])
    binstd_g[i] = np.std(pa_g_sim[col])

numbin = 15
bins_set = np.linspace(0, 360, numbin) - 360 / numbin / 2

fig, (ax0, ax1, ax2) = plt.subplots(figsize=(6, 6), nrows=3, sharex=True)

ax0.plot(bins_set[1:], binmean_k, "r-", lw=1)
ax0.fill_between(bins_set[1:], binmean_k+binstd_k, binmean_k-binstd_k,
                 color="k", alpha=0.1)
ax0.fill_between(bins_set[1:], binmean_k+2*binstd_k, binmean_k+binstd_k,
                 color="g", alpha=0.1)
ax0.fill_between(bins_set[1:], binmean_k-binstd_k, binmean_k-2*binstd_k,
                 color="g", alpha=0.1)
ax0.fill_between(bins_set[1:], binmean_k+3*binstd_k, binmean_k+2*binstd_k,
                 color="r", alpha=0.1)
ax0.fill_between(bins_set[1:], binmean_k-3*binstd_k, binmean_k-2*binstd_k,
                 color="r", alpha=0.1)

ax1.plot(bins_set[1:], binmean_ka, "r-", lw=1)
ax1.fill_between(bins_set[1:], binmean_ka+binstd_ka, binmean_ka-binstd_ka,
                 color="k", alpha=0.1)
ax1.fill_between(bins_set[1:], binmean_ka+2*binstd_ka, binmean_ka+binstd_ka,
                 color="g", alpha=0.1)
ax1.fill_between(bins_set[1:], binmean_ka-binstd_ka, binmean_ka-2*binstd_ka,
                 color="g", alpha=0.1)
ax1.fill_between(bins_set[1:], binmean_ka+3*binstd_ka, binmean_ka+2*binstd_ka,
                 color="r", alpha=0.1)
ax1.fill_between(bins_set[1:], binmean_ka-3*binstd_ka, binmean_ka-2*binstd_ka,
                 color="r", alpha=0.1)

ax2.plot(bins_set[1:], binmean_g, "r-", lw=1)
ax2.fill_between(bins_set[1:], binmean_g+binstd_g, binmean_g-binstd_g,
                 color="k", alpha=0.1)
ax2.fill_between(bins_set[1:], binmean_g+2*binstd_g, binmean_g+binstd_g,
                 color="g", alpha=0.1)
ax2.fill_between(bins_set[1:], binmean_g-binstd_g, binmean_g-2*binstd_g,
                 color="g", alpha=0.1)
ax2.fill_between(bins_set[1:], binmean_g+3*binstd_g, binmean_g+2*binstd_g,
                 color="r", alpha=0.1)
ax2.fill_between(bins_set[1:], binmean_g-3*binstd_g, binmean_g-2*binstd_g,
                 color="r", alpha=0.1)

ax0.set_xlim([0, 360])

ax0.text(40, 55, "$PA_{\\rm X-K}$", fontsize=12)
ax1.text(40, 70, "$PA_{\\rm X-Ka}$", fontsize=12)
ax2.text(40, 50, "$PA_{\\rm X-G}$", fontsize=12)

ax2.set_xticks(np.arange(0, 361, 60))
ax2.xaxis.set_minor_locator(MultipleLocator(30))

ax2.set_xlabel("PA")
ax0.set_ylabel("No. Sources")
ax1.set_ylabel("No. Sources")
ax2.set_ylabel("No. Sources")

plt.tight_layout()
plt.subplots_adjust(hspace=0)
```

The peak around zero of differences amongst PA(X-to-K), PA(X-to-Ka), and PA(X-to-Gaia) could also be recovered, significant at three-sigma level.

```python
pa_k_g_sim = Table.read("../data/pa_k_gaia_sim_hist.dat", format="ascii")
pa_ka_g_sim = Table.read("../data/pa_ka_gaia_sim_hist.dat", format="ascii")
pa_k_ka_sim = Table.read("../data/pa_k_ka_sim_hist.dat", format="ascii")
```

```python
cols = pa_k_g_sim.colnames

binmean_k_g = np.zeros(len(cols))
binstd_k_g = np.zeros(len(cols))
binmean_k_ka = np.zeros(len(cols))
binstd_k_ka = np.zeros(len(cols))
binmean_ka_g = np.zeros(len(cols))
binstd_ka_g = np.zeros(len(cols))

for i, col in enumerate(cols):
    binmean_k_g[i] = np.mean(pa_k_g_sim[col])
    binstd_k_g[i] = np.std(pa_k_g_sim[col])

    binmean_ka_g[i] = np.mean(pa_ka_g_sim[col])
    binstd_ka_g[i] = np.std(pa_ka_g_sim[col])

    binmean_k_ka[i] = np.mean(pa_k_ka_sim[col])
    binstd_k_ka[i] = np.std(pa_k_ka_sim[col])
```

```python
# Interpolate
numbin = 15
node = np.linspace(-180, 180, numbin)
bins_node = node[1:] - 360 / numbin / 2

# X-to-K - X-to-Gaia
# Mean
fm1 = interpolate.UnivariateSpline(bins_node, binmean_k_g, s=0)
binmean_k_g1 = fm1(node)
# Std
fm1 = interpolate.UnivariateSpline(bins_node, binstd_k_g, s=0)
binstd_k_g1 = fm1(node)

# X-to-K - X-to-Gaia
# Mean
fm1 = interpolate.UnivariateSpline(bins_node, binmean_k_ka, s=0)
binmean_k_ka1 = fm1(node)
# Std
fm1 = interpolate.UnivariateSpline(bins_node, binstd_k_ka, s=0)
binstd_k_ka1 = fm1(node)

# X-to-Ka - X-to-Gaia
# Mean
fm1 = interpolate.UnivariateSpline(bins_node, binmean_ka_g, s=0)
binmean_ka_g1 = fm1(node)
# Std
fm1 = interpolate.UnivariateSpline(bins_node, binstd_ka_g, s=0)
binstd_ka_g1 = fm1(node)
```

```python
fig, (ax0, ax1, ax2) = plt.subplots(figsize=(6, 6), nrows=3, sharex=True)

ax0.plot(node, binmean_k_g1, "r-", lw=1)
ax0.fill_between(node, binmean_k_g1+binstd_k_g1, binmean_k_g1-binstd_k_g1,
                 color="r", alpha=0.2)
ax0.fill_between(node, binmean_k_g1+2*binstd_k_g1, binmean_k_g1+binstd_k_g1,
                 color="y", alpha=0.2)
ax0.fill_between(node, binmean_k_g1-binstd_k_g1, binmean_k_g1-2*binstd_k_g1,
                 color="y", alpha=0.2)
ax0.fill_between(node, binmean_k_g1+3*binstd_k_g1, binmean_k_g1+2*binstd_k_g1,
                 color="g", alpha=0.2)
ax0.fill_between(node, binmean_k_g1-3*binstd_k_g1, binmean_k_g1-2*binstd_k_g1,
                 color="g", alpha=0.2)

ax1.plot(node, binmean_k_ka1, "r-", lw=1)
ax1.fill_between(node, binmean_k_ka1+binstd_k_ka1, binmean_k_ka1-binstd_k_ka1,
                 color="r", alpha=0.2)
ax1.fill_between(node, binmean_k_ka1+2*binstd_k_ka1, binmean_k_ka1+binstd_k_ka1,
                 color="y", alpha=0.2)
ax1.fill_between(node, binmean_k_ka1-binstd_k_ka1, binmean_k_ka1-2*binstd_k_ka1,
                 color="y", alpha=0.2)
ax1.fill_between(node, binmean_k_ka1+3*binstd_k_ka1, binmean_k_ka1+2*binstd_k_ka1,
                 color="g", alpha=0.2)
ax1.fill_between(node, binmean_k_ka1-3*binstd_k_ka1, binmean_k_ka1-2*binstd_k_ka1,
                 color="g", alpha=0.2)

ax2.plot(node, binmean_ka_g1, "r-", lw=1)
ax2.fill_between(node, binmean_ka_g1+binstd_ka_g1, binmean_ka_g1-binstd_ka_g1,
                 color="r", alpha=0.2)
ax2.fill_between(node, binmean_ka_g1+2*binstd_ka_g1, binmean_ka_g1+binstd_ka_g1,
                 color="y", alpha=0.2)
ax2.fill_between(node, binmean_ka_g1-binstd_ka_g1, binmean_ka_g1-2*binstd_ka_g1,
                 color="y", alpha=0.2)
ax2.fill_between(node, binmean_ka_g1+3*binstd_ka_g1, binmean_ka_g1+2*binstd_ka_g1,
                 color="g", alpha=0.2)
ax2.fill_between(node, binmean_ka_g1-3*binstd_ka_g1, binmean_ka_g1-2*binstd_ka_g1,
                 color="g", alpha=0.2)

ax0.text(-150, 60, "$PA_{\\rm X-K} - PA_{\\rm X-G}$", fontsize=12)
ax1.text(-150, 80, "$PA_{\\rm X-K} - PA_{\\rm X-Ka}$", fontsize=12)
ax2.text(-150, 60, "$PA_{\\rm X-Ka} - PA_{\\rm X-G}$", fontsize=12)

ax0.set_xlim([-180, 180])

ax2.set_xticks(np.arange(-180, 181, 60))
ax2.xaxis.set_minor_locator(MultipleLocator(30))

ax2.set_xlabel("Simulated PA difference (${}^{\circ}$)")
ax0.set_ylabel("No. Sources")
ax1.set_ylabel("No. Sources")
ax2.set_ylabel("No. Sources")

ax0.grid(lw=0.1)
ax1.grid(lw=0.1)
ax2.grid(lw=0.1)

plt.tight_layout()
plt.subplots_adjust(hspace=0)
```
