```python
from astropy.table import Table, join, Column
from astropy import units as u
import astropy.coordinates as coord
from astropy.coordinates import SkyCoord
import numpy as np

import matplotlib.pyplot as plt
from matplotlib.ticker import MultipleLocator, NullFormatter
%matplotlib inline
```

```python
r2o= Table.read("../data/common-source-position-wrt-gaia.fits")
```

```python
# Gaia DR2 photometric information
from my_progs.catalog.read_gaia import read_dr2_iers

gdr2 = read_dr2_iers()

gdr2.keep_columns(["iers_name", "source_id",
                   "ra", "ra_err", "dec", "dec_err", "pos_err",
                   "parallax", "parallax_err",
                   "pmra", "pmra_err", "pmdec", "pmdec_err",
                   "astrometric_n_obs_al", "astrometric_matched_observations",
                   "phot_g_mean_mag", "phot_bp_mean_mag", "phot_rp_mean_mag", "bp_rp"])

gdr2.rename_columns(["astrometric_n_obs_al", "astrometric_matched_observations",
                     "phot_g_mean_mag", "phot_bp_mean_mag", "phot_rp_mean_mag"],
                    ["n_obs_al", "n_matched_obs", "g_mag", "bp_mag", "rp_mag"])

r2o = join(r2o, gdr2, keys="iers_name", table_names=["", "g"])
```

I plot the distribution of the $G$-mag in order to determine the max and min value of colorbar.

```python
# Distribution of Gaia error
print("The range of Gaia G-mag is between {:.3f} and {:.3f}".format(
    np.min(r2o["g_mag"]), np.max(r2o["g_mag"])))

bins = np.arange(12.5, 21.1, 0.5)
plt.hist(r2o["g_mag"], bins)
```

```python
# Distribution of Gaia error
print("The range of Gaia BP-RP is between {:.3f} and {:.3f}".format(
    np.min(r2o["bp_rp"]), np.max(r2o["bp_rp"])))

plt.hist(r2o["bp_rp"])
```

# Angular separation $\rho$

```python
# Distribution of angular separation
fig, (ax0, ax1, ax2) = plt.subplots(figsize=(6, 8), nrows=3, sharex=True, sharey=True)

sc0 = ax0.scatter(r2o["pos_err"], r2o["pos_err_x"], c=r2o["g_mag"],
#                   s=1, vmin=16, vmax=20, cmap="Greys")
                  s=1, vmin=16, vmax=20, cmap="coolwarm")
plt.colorbar(sc0, label="$G$ (mag)", fraction=0.1, ax=ax0)

sc1 = ax1.scatter(r2o["pos_err"], r2o["pos_err_k"], c=r2o["g_mag"],
#                   s=1, vmin=16, vmax=20, cmap="Greys")
                  s=1, vmin=16, vmax=20, cmap="coolwarm")
plt.colorbar(sc1, label="$G$ (mag)", fraction=0.1, ax=ax1)

sc2 = ax2.scatter(r2o["pos_err"], r2o["pos_err_ka"], c=r2o["g_mag"],
#                   s=1, vmin=16, vmax=20, cmap="Greys")
                  s=1, vmin=16, vmax=20, cmap="coolwarm")
plt.colorbar(sc2, label="$G$ (mag)", fraction=0.1, ax=ax2)


ax0.set_ylabel("EEMA (X, mas)")
ax1.set_ylabel("EEMA (K, mas)")
ax2.set_ylabel("EEMA (Ka, mas)")
ax2.set_xlabel("EEMA (Gaia, mas)")

ax0.set_xscale("log")
ax1.set_xscale("log")
ax2.set_xscale("log")
ax0.set_yscale("log")
ax1.set_yscale("log")
ax2.set_yscale("log")

ax0.xaxis.set_ticks_position("both")
ax0.yaxis.set_ticks_position("both")
ax1.xaxis.set_ticks_position("both")
ax1.yaxis.set_ticks_position("both")
ax2.xaxis.set_ticks_position("both")
ax2.yaxis.set_ticks_position("both")

ax0.set_xlim([0.01, 2.5])

plt.subplots_adjust()
plt.tight_layout()
```

This plot shows that the VLBI position precision has less correlation with the $G$-magnitude.

# What do we find?

- 1. VLBI position formal uncertainty is less relevant to the $Gaia$ $G$-mag.

- 2. No evidence shows that the K-Gaia and Ka-Gaia distance is smaller than the X-Gaia distance.
Radio source structure index information is required for further comparison.

- 3. The radio-to-optical distance at three bands slightly derease as moving towards to the faint end.

- 4. No clear dependency of the distribution of angular and normalized separation on the $G$-mag or 
$G_{bp}-G_{rp}$ is observed
