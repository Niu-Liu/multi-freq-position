With the ICRF3 and Gaia DR2 catalog now we have the quasar position measured indepeddently at four different bands:
S/X, K, and X/Ka from VLBI and optical one from Gaia DR2.

Comparing these four positions may be interesting for studying the wavalength-dependency of quasar position and astrophysical properties of AGNs.

Here I use the S/X position as the reference position (assume this position is the most accurate) and compute the position offsets of other three bands.

Several perspects are investigated in this note, as listed below.

- dependency of positional offset on the R.A. and decl.;
- statistics of angular separation (denoted as $\rho$) and normalized separation (denoted as $X$);
- statistical correlation amongst position angles (PAs) of positional offset vector;
- statistical correlation between position angles and other quantity, such as the direction of Gaia DR2 proper motion, direction of major axis of error ellipses;

This notebook show the first part of this work, where I did NOT remove the global systematical difference amongst 
catalogs before my analysis.

## 1. Multi-wavelength positional offsets
### 1.1 Calculation

```python
from astropy.table import Table, join, Column
from astropy import units as u
import numpy as np

import matplotlib.pyplot as plt
from matplotlib.ticker import MultipleLocator, NullFormatter
%matplotlib inline
```

Get the quasar positions from 4 catalogs and cross-match them to find the common sources.

```python
from my_progs.catalog import read_icrf, read_gaia

icrf3sx = read_icrf.read_icrf3(wv="sx")
icrf3k = read_icrf.read_icrf3(wv="k")
icrf3xka = read_icrf.read_icrf3(wv="xka")

gaiadr2 = read_gaia.read_dr2_iers()
```

```python
tmp1 = join(icrf3sx, icrf3k, keys=["icrf_name", "iers_name"], table_names=["x", "k"])
tmp2 = join(icrf3xka, gaiadr2, keys="iers_name", table_names=["ka", "g"])

comsou = join(tmp1, tmp2, keys=["icrf_name", "iers_name"])

print("We obtain {:d} common sources amongst 4 catalogs.".format(len(comsou)))
```

```python
from my_progs.catalog.pos_diff import radio_cat_diff_calc
# Calculate the positional offset wrt. X position
# K - X
k2x = radio_cat_diff_calc(icrf3k, icrf3sx, "iers_name", ["k", "x"])

# Ka - X
ka2x = radio_cat_diff_calc(icrf3xka, icrf3sx, "iers_name", ["ka", "x"])
ka2x.remove_columns(["ra", "dec", "ra_err_x", "dec_err_x", "pos_err_x"])

# Gaia - X
g2x = radio_cat_diff_calc(gaiadr2, icrf3sx, "iers_name", ["g", "x"])
g2x.remove_columns(["ra", "dec", "ra_err_x", "dec_err_x", "pos_err_x"])
```

```python
comsou = join(k2x, ka2x, keys="iers_name", table_names=["k", "ka"])
comsou = join(comsou, g2x, keys="iers_name")

# Rename some columns
comsou.rename_columns(["dra", "ddec", "dra_err", "ddec_err", "dra_ddec_cov",
                       "ang_sep", "nor_ra", "nor_dec", "nor_sep"],
                      ["dra_g", "ddec_g", "dra_err_g", "ddec_err_g",
                       "dra_ddec_cov_g",
                       "ang_sep_g", "nor_ra_g", "nor_dec_g", "nor_sep_g"])


# Add proper motion information
gdr2pm = Table(gaiadr2)
gdr2pm.keep_columns(["iers_name", "pmra", "pmra_err", "pmdec", "pmdec_err", "ra_dec_corr"])
gdr2pm.rename_column("ra_dec_corr", "ra_dec_corr_g")

comsou = join(comsou, gdr2pm, keys="iers_name")

# Add correlation coefficent
corrsx = Table(icrf3sx)
corrsx.keep_columns(["iers_name", "ra_dec_corr"])
corrsx.rename_column("ra_dec_corr", "ra_dec_corr_x")

corrk = Table(icrf3k)
corrk.keep_columns(["iers_name", "ra_dec_corr"])
corrk.rename_column("ra_dec_corr", "ra_dec_corr_k")

corrxka = Table(icrf3xka)
corrxka.keep_columns(["iers_name", "ra_dec_corr"])
corrxka.rename_column("ra_dec_corr", "ra_dec_corr_ka")

comsou = join(comsou, corrsx, keys="iers_name")
comsou = join(comsou, corrk, keys="iers_name")
comsou = join(comsou, corrxka, keys="iers_name")
```

There are 488 common sources in the four catalogs.

I use the S/X-band position as the reference point and calculate the positional offset vector of K-band, X/Ka-band, and Gaia, 
as illustrated in the following diagram.

Please note in the plot $\Delta\alpha\cos\delta$ decreases from the left to right.
This kind of convention is used in the VLBI imaging so I follow the same convention for further comparison.

From the position offset vector I calculate the angular separation $\rho$ 
and normalized separation $X$ as done in the Mignard et 2018. 

```python
# Positional offset vector
fig, ax = plt.subplots()

souname = "2134+004"
mask = (comsou["iers_name"] == souname)
comsou1 = comsou[mask]

ax.plot(0, 0, "r+", label="X", ms=10)
ax.plot(comsou1[0]["dra_k"], comsou1[0]["ddec_k"], "ms", label="K")
ax.plot(comsou1[0]["dra_ka"], comsou1[0]["ddec_ka"], "bv", label="X/Ka")
ax.plot(comsou1[0]["dra_g"], comsou1[0]["ddec_g"], "y^", label="Gaia")

ax.arrow(0, 0, comsou1[0]["dra_k"], comsou1[0]["ddec_k"], color="m")
ax.arrow(0, 0, comsou1[0]["dra_ka"], comsou1[0]["ddec_ka"], color="b")
ax.arrow(0, 0, comsou1[0]["dra_g"], comsou1[0]["ddec_g"], color="y")

# TO DO
# Plot error ellipse

# ax.set_xlim(0.5, -0.2)

ax.set_xlabel("$\Delta\,\\alpha\,\cos\delta$ (mas)")
ax.set_ylabel("$\Delta\,\delta$ (mas)")
ax.set_title(comsou1[0]["iers_name"])

# ax.grid()
ax.legend()
```

## Positional angle of position offset vector

The position angle, which characterize direction of position offset vector, is measured as angle between the direction of increasing-declination and position offset vector clockwisely.

As a result, the position angles of K-band, X/Ka-band, and Gaia position wrt. the S/X band one for 0002+541 are $217^\circ$, $135^\circ$, $63^\circ$, respectively.

```python
from calc_pa import pa_calc

# Calculate the positional angle of the positional offset
PA_k = pa_calc(comsou["dra_k"], comsou["ddec_k"])
PA_ka = pa_calc(comsou["dra_ka"], comsou["ddec_ka"])
PA_g = pa_calc(comsou["dra_g"], comsou["ddec_g"])

# Add these parameters into the data table as new columns
PA_k = Column(PA_k, name="pa_k_x", unit=u.deg)
PA_ka = Column(PA_ka, name="pa_ka_x", unit=u.deg)
PA_g = Column(PA_g, name="pa_g_x", unit=u.deg)

comsou.add_columns([PA_k, PA_ka, PA_g])
```

The diagram below despicts the distribution of positional angles for K-band, X/Ka-band, and Gaia position wrt. S/X-band position.

A steep peak aroud $180^\circ$ is seen in X/Ka$-$S/X, most likely caused by the negitive declination bias seen in plots above.

There is no favorable direction of K$-$S/X and Gaia$-$S/X.

```python
# Histogram of angular separation
fig, (ax0, ax1, ax2) = plt.subplots(figsize=(8, 6), nrows=3, sharex=True) # , sharey=True

bins_set = np.linspace(0, 360, 50)

# Weights to calculate the per cent
wgt = 100 / len(comsou) * np.ones_like(comsou["ra"])

# Plot
# K -X
ax0.hist(PA_k, bins_set, histtype="step", fill=False, weights=wgt)

# Ka -X
ax1.hist(PA_ka, bins_set, histtype="step", fill=False, weights=wgt)

# Gaia -X
ax2.hist(PA_g, bins_set, histtype="step", fill=False, weights=wgt)

ax0.set_ylabel("% (K $-$ X)", fontsize=12)
ax1.set_ylabel("% (Ka $-$ X)", fontsize=12)
ax2.set_ylabel("% (Gaia $-$ X)", fontsize=12)
ax2.set_xlabel("Position Angle ($^\circ$)", fontsize=12)
ax2.set_xticks(np.arange(0, 361, 30))

ax0.xaxis.set_ticks_position("both")
ax0.yaxis.set_ticks_position("both")
ax1.xaxis.set_ticks_position("both")
ax1.yaxis.set_ticks_position("both")
ax2.xaxis.set_ticks_position("both")
ax2.yaxis.set_ticks_position("both")

ax2.xaxis.set_minor_locator(MultipleLocator(10))

# plt.savefig("../plots/pa-hist0.eps")
```

I calculate the differences amongst positional angle of positional offset.

If the positional offset is negative, I add $360^\circ$ to make sure that the positional offset ranges from 0 to $360^\circ$.

The distribution of position angle offset is shown below.

We can find that the position angle offset between K$-$S/X and Gaia$-$S/X vector peaks at 0, 
indicating that the K-band and Gaia positions locates in nearly same direction as seen from the S/X position.

In the following figure, I directly look at the angle between positional offset vectors.

```python
from calc_pa import inclu_angle_calc

# Calculate the angle between positional offset vectors
ang_k_g = inclu_angle_calc(
    comsou["dra_k"], comsou["ddec_k"], comsou["dra_g"], comsou["ddec_g"])
ang_ka_g = inclu_angle_calc(
    comsou["dra_ka"], comsou["ddec_ka"], comsou["dra_g"], comsou["ddec_g"])
ang_ka_k = inclu_angle_calc(
    comsou["dra_k"], comsou["ddec_k"], comsou["dra_ka"], comsou["ddec_ka"])

# Add these parameters into the data table as new columns
ang_k_g = Column(ang_k_g, name="inc_ang_k_g", unit=u.deg)
ang_ka_g = Column(ang_ka_g, name="inc_ang_ka_g", unit=u.deg)
ang_ka_k = Column(ang_ka_k, name="inc_ang_ka_k", unit=u.deg)

comsou.add_columns([ang_k_g, ang_ka_g, ang_ka_k])
```

The included angles between X/Ka and K shows a peak at $180^\circ$ and it is about $10^\circ$ between K and Gaia.

It implys that the positional offset vectors of K$-$S/X and Gaia$-$S/X tend to point towards the same direction,
while positional offset vectors of K$-$S/X and X/Ka$-$S/X is likely to point towards the oppisite direction.

If so, we could expect a peak of $180^\circ$ between Gaia$-$S/X and X/Ka$-$S/X.
However, the distribution of included angle between Gaia$-$S/X and X/Ka$-$S/X is flat.

```python
# Histogram of angular separation
fig, (ax0, ax1, ax2) = plt.subplots(figsize=(8, 6), nrows=3, sharex=True) # , sharey=True

bins_set = np.linspace(0, 180, 50)

# Weights to calculate the per cent
wgt = 100 / len(comsou) * np.ones_like(comsou["ra"])

# Plot
# K -X
ax0.hist(ang_k_g, bins_set, histtype="step", fill=False, weights=wgt)

# Ka -X
ax1.hist(ang_ka_g, bins_set, histtype="step", fill=False, weights=wgt)

# Gaia -X
ax2.hist(ang_ka_k, bins_set, histtype="step", fill=False, weights=wgt)

ax0.set_ylabel("% (K-X and Gaia-X)", fontsize=8)
ax1.set_ylabel("% (Ka-X and Gaia-X)", fontsize=8)
ax2.set_ylabel("% (Ka-X and K-X)", fontsize=8)
ax2.set_xlabel("Angle between position offset vectors ($^\circ$)", fontsize=12)
ax2.set_xticks(np.arange(0, 181, 30))

ax2.xaxis.set_minor_locator(MultipleLocator(10))

# plt.savefig("../plots/pa-diff0.eps")
```

To check if the position angle of positional offset vector is biased by the origin position formal error at each band, I calculate the position angle of the direction of major axis of error ellipse (EEMA) for each catalog.


```python
from calc_pa import eepa_calc

# Calculate the positional angle of the major axis of error ellipse
eepa_x = eepa_calc(comsou["ra_err_x"], comsou["dec_err_x"], comsou["ra_dec_corr_x"])
eepa_k = eepa_calc(comsou["ra_err_k"], comsou["dec_err_k"], comsou["ra_dec_corr_k"])
eepa_ka = eepa_calc(comsou["ra_err_ka"], comsou["dec_err_ka"], comsou["ra_dec_corr_ka"])
eepa_g = eepa_calc(comsou["ra_err_g"], comsou["dec_err_g"], comsou["ra_dec_corr_g"])

# Add these parameters into the data table as new columns
eepa_x = Column(eepa_x, name="eepa_x", unit=u.deg)
eepa_k = Column(eepa_k, name="eepa_k", unit=u.deg)
eepa_ka = Column(eepa_ka, name="eepa_ka", unit=u.deg)
eepa_g = Column(eepa_g, name="eepa_g", unit=u.deg)

comsou.add_columns([eepa_x, eepa_k, eepa_ka, eepa_g])
```

The error ellipse from VLBI catalogs (S/X, K, and X/Ka) has a priviledge direction of $\sim180^\circ$, especially for K-band, which can be expected.

```python
# Histogram of position angle of eema
fig, (ax0, ax1, ax2, ax3) = plt.subplots(figsize=(6, 6), nrows=4, sharex=True)

bins_set = np.linspace(0, 180, 50)

# Weights to calculate the per cent
wgt = 100 / len(comsou) * np.ones_like(comsou["ra"])

# Plot
ax0.hist(eepa_x, bins_set, histtype="step", fill=False, color="k", weights=wgt)
ax1.hist(eepa_k, bins_set, histtype="step", fill=False, color="m", weights=wgt)
ax2.hist(eepa_ka, bins_set, histtype="step", fill=False, color="b", weights=wgt)
ax3.hist(eepa_g, bins_set, histtype="step", fill=False, color="y", weights=wgt)


ax0.set_ylabel("% ", fontsize=12)
ax1.set_ylabel("%", fontsize=12)
ax2.set_ylabel("%", fontsize=12)
ax3.set_ylabel("%", fontsize=12)

ax3.set_xlabel("Position angle of EEMA ($^\circ$)", fontsize=12)
ax3.xaxis.set_minor_locator(MultipleLocator(10))
ax3.set_xticks(np.arange(0, 181, 30))

ax0.text(10, 8, "X", color="k", fontsize=15)
ax1.text(10, 20, "K", color="m", fontsize=15)
ax2.text(10, 12, "Ka", color="b", fontsize=15)
ax3.text(10, 3, "Gaia", color="y", fontsize=15)

plt.tight_layout()
# plt.savefig("../plots/eema-pa-hist.eps")
```
