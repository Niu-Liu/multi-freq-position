```python
# Calculate the included angle
inc_ang_k_g1 = wrap_angle(pa_k_x1 - pa_g_x1)
# inc_ang_k_g1 = Column(inc_ang_k_g1, name="inc_ang_k_g_all2", unit=u.deg)
inc_ang_k_g1 = Column(inc_ang_k_g1, name="inc_ang_k_g_ccl2", unit=u.deg)

inc_ang_ka_k1 = wrap_angle(pa_k_x1 - pa_ka_x1)
# inc_ang_ka_k1 = Column(inc_ang_ka_k1, name="inc_ang_ka_k_all2", unit=u.deg)
inc_ang_ka_k1 = Column(inc_ang_ka_k1, name="inc_ang_k_ka_ccl2", unit=u.deg)

inc_ang_ka_g1 = wrap_angle(pa_g_x1 - pa_ka_x1)
# inc_ang_ka_g1 = Column(inc_ang_ka_g1, name="inc_ang_ka_g_all2", unit=u.deg)
inc_ang_ka_g1 = Column(inc_ang_ka_g1, name="inc_ang_ka_g_ccl2", unit=u.deg)

comsou.add_columns([inc_ang_k_g1, inc_ang_ka_k1, inc_ang_ka_g1])
```

```python
# PA(X-K) - PA(X-Gaia)
pa_k_g_sim = Table.read("../data/sim-as-seb/pa_k_gaia_sim_hist_1.dat", format="ascii")
cols = pa_k_g_sim.colnames

binmean_k_g = np.zeros(len(cols))
binstd_k_g = np.zeros(len(cols))

for i, col in enumerate(cols):
    binmean_k_g[i] = np.mean(pa_k_g_sim[col])
    binstd_k_g[i] = np.std(pa_k_g_sim[col])
    
# PA(X-K) - PA(X-Ka)
pa_k_ka_sim = Table.read("../data/sim-as-seb/pa_k_ka_sim_hist_1.dat", format="ascii")
cols = pa_k_ka_sim.colnames

binmean_k_ka = np.zeros(len(cols))
binstd_k_ka = np.zeros(len(cols))

for i, col in enumerate(cols):
    binmean_k_ka[i] = np.mean(pa_k_ka_sim[col])
    binstd_k_ka[i] = np.std(pa_k_ka_sim[col])
    
# PA(X-Ka) - PA(X-Gaia)
pa_ka_g_sim = Table.read("../data/sim-as-seb/pa_ka_gaia_sim_hist_1.dat", format="ascii")
cols = pa_ka_g_sim.colnames
binmean_ka_g = np.zeros(len(cols))
binstd_ka_g = np.zeros(len(cols))

for i, col in enumerate(cols):
    binmean_ka_g[i] = np.mean(pa_ka_g_sim[col])
    binstd_ka_g[i] = np.std(pa_ka_g_sim[col])
```

```python
# Interpolate
numbin = 15
node = np.linspace(-180, 180, numbin)
bins_node = node[1:] - 360 / numbin / 2

# X-to-K - X-to-Gaia
# Mean
fm1 = interpolate.UnivariateSpline(bins_node, binmean_k_g, s=0)
binmean_k_g1 = fm1(node)
# Std
fm1 = interpolate.UnivariateSpline(bins_node, binstd_k_g, s=0)
binstd_k_g1 = fm1(node)

# X-to-K - X-to-Gaia
# Mean
fm1 = interpolate.UnivariateSpline(bins_node, binmean_k_ka, s=0)
binmean_k_ka1 = fm1(node)
# Std
fm1 = interpolate.UnivariateSpline(bins_node, binstd_k_ka, s=0)
binstd_k_ka1 = fm1(node)

# X-to-Ka - X-to-Gaia
# Mean
fm1 = interpolate.UnivariateSpline(bins_node, binmean_ka_g, s=0)
binmean_ka_g1 = fm1(node)
# Std
fm1 = interpolate.UnivariateSpline(bins_node, binstd_ka_g, s=0)
binstd_ka_g1 = fm1(node)
```

```python
# Histogram of angular separation
fig, (ax0, ax1, ax2) = plt.subplots(figsize=(6, 6), nrows=3, sharex=True)
# , sharey=True

# Simulation results
ax0.plot(node, binmean_k_g1, "r-", lw=1)
ax0.fill_between(node, binmean_k_g1+binstd_k_g1, binmean_k_g1-binstd_k_g1,
                 color="lightsalmon")
ax0.fill_between(node, binmean_k_g1+2*binstd_k_g1, binmean_k_g1+binstd_k_g1,
                 color="lightyellow")
ax0.fill_between(node, binmean_k_g1-binstd_k_g1, binmean_k_g1-2*binstd_k_g1,
                 color="lightyellow")
ax0.fill_between(node, binmean_k_g1+3*binstd_k_g1, binmean_k_g1+2*binstd_k_g1,
                 color="lightgreen")
ax0.fill_between(node, binmean_k_g1-3*binstd_k_g1, binmean_k_g1-2*binstd_k_g1,
                 color="lightgreen")

ax1.plot(node, binmean_k_ka1, "r-", lw=1)
ax1.fill_between(node, binmean_k_ka1+binstd_k_ka1, binmean_k_ka1-binstd_k_ka1,
                 color="lightsalmon")
ax1.fill_between(node, binmean_k_ka1+2*binstd_k_ka1, binmean_k_ka1+binstd_k_ka1,
                 color="lightyellow")
ax1.fill_between(node, binmean_k_ka1-binstd_k_ka1, binmean_k_ka1-2*binstd_k_ka1,
                 color="lightyellow")
ax1.fill_between(node, binmean_k_ka1+3*binstd_k_ka1, binmean_k_ka1+2*binstd_k_ka1,
                 color="lightgreen")
ax1.fill_between(node, binmean_k_ka1-3*binstd_k_ka1, binmean_k_ka1-2*binstd_k_ka1,
                 color="lightgreen")

ax2.plot(node, binmean_ka_g1, "r-", lw=1)
ax2.fill_between(node, binmean_ka_g1+binstd_ka_g1, binmean_ka_g1-binstd_ka_g1,
                 color="lightsalmon", step="pre")
ax2.fill_between(node, binmean_ka_g1+2*binstd_ka_g1, binmean_ka_g1+binstd_ka_g1,
                 color="lightyellow")
ax2.fill_between(node, binmean_ka_g1-binstd_ka_g1, binmean_ka_g1-2*binstd_ka_g1,
                 color="lightyellow")
ax2.fill_between(node, binmean_ka_g1+3*binstd_ka_g1, binmean_ka_g1+2*binstd_ka_g1,
                 color="lightgreen")
ax2.fill_between(node, binmean_ka_g1-3*binstd_ka_g1, binmean_ka_g1-2*binstd_ka_g1,
                 color="lightgreen")

bins_set = np.linspace(-180, 180, 13)

# Weights to calculate the per cent
wgt = 100 / len(comsou) * np.ones_like(comsou["ra"])

# Plot
# K -X
ax0.hist(inc_ang_k_g1, bins_set, histtype="step", fill=False, color="b")

# Ka -X
ax1.hist(inc_ang_ka_k1, bins_set, histtype="step", fill=False, color="b")

# Gaia -X
ax2.hist(inc_ang_ka_g1, bins_set, histtype="step", fill=False, color="b")

ax0.set_ylabel("No. sources", fontsize=12)
ax1.set_ylabel("No. sources", fontsize=12)
ax2.set_ylabel("No. sources", fontsize=12)
ax2.set_xlabel("PA difference ($^\circ$)", fontsize=12)
ax2.set_xticks(np.arange(-180, 181, 60))

ax0.text(-150, 60, "$PA_{\\rm XK} - PA_{\\rm XG}$", fontsize=12)
ax1.text(-150, 80, "$PA_{\\rm XK} - PA_{\\rm XKa}$", fontsize=12)
ax2.text(-150, 60, "$PA_{\\rm XKa} - PA_{\\rm XG}$", fontsize=12)

ax2.set_xlim([-180, 180])

ax2.xaxis.set_minor_locator(MultipleLocator(30))

# ax0.grid(lw=0.1)
# ax1.grid(lw=0.1)
# ax2.grid(lw=0.1)

plt.tight_layout()
# plt.savefig("../plots/pa-diff.eps", hbox="tight")
```
